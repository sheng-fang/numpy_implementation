```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
class KalmanFilter():
    def __init__(self, F, H, B=None, R=None, P=None, x0=None):
        self.n = F.shape[1]
        self.m = H.shape[1]

        self.F = F
        self.H = H
        if B is None:
            self.B = 0  
        else:
            self.B = B
            
        if R is None:
            self.R = np.eye(self.n)
        else:
            self.R = R
            
        if P is None:
            self.P = np.eye(self.n) * 1000
        else:
            self.P = P
        
        if x0 is None:
            self.x = np.zeros((self.n, 1)) 
        else:
            self.x = x0

    def predict(self):
        self.x = np.dot(self.F, self.x)
        self.P = np.dot(np.dot(self.F, self.P), self.F.T) + self.B
        return self.x

    def update(self, z):
        y = z - np.dot(self.H, self.x)
        S = np.dot(self.H, np.dot(self.P, self.H.T)) + self.R
        K = np.dot(np.dot(self.P, self.H.T), np.linalg.inv(S))
        self.x = self.x + np.dot(K, y)
        I = np.eye(self.n)
        self.P = np.dot(I - np.dot(K, self.H), self.P)
```

```python
# init parameters
dt = 1 # observation interval
F = np.array([[1, dt], [0, 1]]).reshape(2, 2)
H = np.array([[1, 0], [0, 0]]).reshape(2, 2)
B = np.random.rand() * 0.001

kf = KalmanFilter(F, H, B=B)
```

```python
# load csv file
with open('pb1.csv') as csv_file:
    file_reader = csv.reader(csv_file)
    obs = []
    for i in file_reader:
        obs.append(i)
```

```python
# apply kalman filter and predict
preds = []
measeure = []
for ob in obs[0]:
    ob_float = float(ob)
    measeure.append(ob_float)
    preds.append(np.dot(H, kf.predict())[0])
    kf.update(np.array([ob_float, kf.x[1][0]]).reshape(2,1))
```
